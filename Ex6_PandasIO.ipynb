O desafio agora é normalizar esse DataFrame, expandindo as colunas address e company em suas respectivas subcolunas para facilitar a visualização e a análise dos dados.



```python
import pandas as pd
```

```python
url = 'https://jsonplaceholder.typicode.com/users'
```

```python
df = pd.read_json(url)
df.head(3)
```

```python
df_address = pd.json_normalize(df['address'])
df_address
```

```python
df_company = pd.json_normalize(df['company'])
df_company
```

```python
df.drop('address', axis=1)
df
```

```python
df_normalizado = pd.concat([df.drop(columns=['address', 'company']), df_address, df_company], axis=1)
df_normalizado
```
